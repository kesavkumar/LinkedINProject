```python
listDetails = []
for i in range(50):
    url = f"https://in.linkedin.com/jobs-guest/jobs/api/seeMoreJobPostings/search?keywords=Data%2BAnalyst&location=Chennai%2C%2BTamil%2BNadu%2C%2BIndia&geoId=106888327&trk=public_jobs_jobs-search-bar_search-submit&start={i}"
    html_data = requests.get(url).text
    soup = BeautifulSoup(html_data,"html.parser")
    job_roleh3=soup.find_all("h3")
    company_nameh4=soup.find_all("h4")
    timeData = soup.find_all("time")
    for index in range(len(company_nameh4)):
        cName = company_nameh4[index].text.strip()
        cRole = job_roleh3[index].text.strip()
        timeD = timeData[index]['datetime']
        listDetails.append([cName,cRole,timeD])
frameDetails = pd.DataFrame(listDetails,columns = ["Company" , "Role" , "Posted_Date"])
```
